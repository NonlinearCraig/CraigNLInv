```python

# Open the text file in read mode with specified encoding
with open('unicycle.txt', 'r', encoding='utf-8') as file:
    # Read the contents of the file
    contents = file.readlines()
for line in contents:
    if "*" in line:
        print(line)
        inv=line.replace('^','**')
        inv_f1=inv.replace('x1','x1_f')
        inv_f1=inv_f1.replace('x2','x2_f')
        
        x1,x2= Reals('x1 x2')
        xvars=[x1,x2]
        inv=eval(inv)

        start = time.time()
        pre = x1**2+(x2-1)**2-1<= 0
        post = x1**2 + (x2-1)**2 - 4<=0
        
        inv = inv>=0
        
        w = 1.0178 + 1.8721 * x1 - 0.0253 * x2
        
        x1_f = x1 + 0.01*(1-x2*w)
        x2_f = x2 + 0.01*x1*w
        inv_f1=eval(inv_f1)
        inv_f1=inv_f1>=0
        loop_cond=1>0
        

        cons1 = ForAll(xvars,Implies(pre, inv))
        cons2 = ForAll(xvars,Implies(And(inv, loop_cond), inv_f1))
        cons3 = ForAll(xvars,Implies(And(inv, Not(loop_cond)),post))

        s = Solver()
        s.add( cons1, cons2, cons3)
        print(s.check())
        if str(s.check())=='sat':
            print(s.model())
        end = time.time()
        print("Consume Time: %.8f s\n"%((end-start)))
```
